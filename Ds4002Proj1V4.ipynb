```python
from google.colab import drive
from PIL import Image
import os

# Mount Google Drive
drive.mount('/content/drive')

# Parameters
testsize = 352
base_path = '/content/drive/My Drive/DS 4002 Project Group 14/'  # Adjust this path to the location of your dataset on Google Drive
dataset_folder = 'Project 3/Data/Camouflaged/'
test_save = os.path.join(base_path, 'Result/2020-CVPR-SINet-New/COD10K/')

# Ensure the save directory exists
os.makedirs(test_save, exist_ok=True)

def load_images(image_root):
    """
    Generator to load images from the specified directory
    """
    for img_name in os.listdir(image_root):
        image_path = os.path.join(image_root, img_name)
        image = Image.open(image_path).convert('RGB')
        yield image, img_name

image_root = os.path.join(base_path, dataset_folder, 'Train/')

img_count = 1
for image, name in load_images(image_root):
    # Resize the image
    resized_image = image.resize((testsize, testsize))

    # Save resized image
    image_save_path = os.path.join(test_save, name)
    resized_image.save(image_save_path)

    print(f'[Image Processed] Dataset: COD10K, Image: {name} ({img_count})')
    img_count += 1

print("\n[Congratulations! Image Processing Done]")


```
